# Test NVIDIA NIM API: nvidia/nemotron-3-nano-omni-30b-a3b-reasoning

Base URL: `https://integrate.api.nvidia.com/v1`  
Docs: https://docs.api.nvidia.com/nim/reference/nvidia-nemotron-3-nano-omni-30b-a3b-reasoning-infer

```python
import os
from dotenv import load_dotenv
from openai import OpenAI

load_dotenv()

MODEL = "nvidia/nemotron-3-nano-omni-30b-a3b-reasoning"

client = OpenAI(
    base_url="https://integrate.api.nvidia.com/v1",
    api_key=os.getenv("NVIDIA_NIM_API_KEY"),
)
```

```python
def chat(prompt: str, model: str = MODEL) -> str:
    """Send a chat prompt to the NVIDIA NIM reasoning model."""
    response = client.chat.completions.create(
        model=model,
        messages=[{"role": "user", "content": prompt}],
        temperature=0.6,
        top_p=0.95,
        max_tokens=65536,
        stream=False,
        extra_body={
            "chat_template_kwargs": {"enable_thinking": True},
            "reasoning_budget": 16384,
        },
    )
    return response.choices[0].message.content
```

```python
# Test 1: simple factual question
reply = chat("What is the capital of France?")
print(reply)
```

```python
# Test 2: reasoning / logic
reply = chat("How many r's are in the word 'strawberry'? Explain your reasoning step-by-step.")
print(reply)
```

```python
# Test 3: creative generation
reply = chat("Write a haiku about artificial intelligence.")
print(reply)
```
