# Test google/gemma-4-31b-it:free via OpenRouter
Simple prompt-response notebook.

```python
import os
from dotenv import load_dotenv
from openai import OpenAI

load_dotenv()

MODEL = "google/gemma-4-31b-it:free"

client = OpenAI(
    base_url="https://openrouter.ai/api/v1",
    api_key=os.getenv("OPENROUTER_API_KEY"),
)
```

```python
def chat(prompt: str, model: str = MODEL) -> str:
    """Send a single-turn chat prompt and return the assistant's reply."""
    response = client.chat.completions.create(
        model=model,
        messages=[{"role": "user", "content": prompt}],
    )
    return response.choices[0].message.content
```

```python
# Test 1: simple factual question
reply = chat("What is the capital of France?")
print(reply)
```

```python
# Test 2: reasoning / logic
reply = chat("How many r's are in the word 'strawberry'? Explain your reasoning.")
print(reply)
```

```python
# Test 3: creative generation
reply = chat("Write a haiku about artificial intelligence.")
print(reply)
```
